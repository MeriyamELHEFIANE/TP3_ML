## *Question 2*

```python
#
import numpy as np
import matplotlib.pyplot as plt

def compute_error_for_line_given_point(b,w,points):
    totalerror=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        totalerror+=(y-(w*x+b))**2
    return totalerror/(2*float(len(points)))
```

## *Question 3*

```python
#
def step_gradent(b_current,w_current,points,learnrate):
    b_gradient=0
    w_gradient=0
    N=float(len(points))

    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        #
        b_gradient+=(1/N)*((w_current*x+b_current)-y)#
        w_gradient+=(1/N)*x*((w_current*x+b_current)-y)
    #renew
    new_b=b_current-(learnrate*b_gradient)
    new_w=w_current-(learnrate*w_gradient)
    return [new_b,new_w]
```

## *Question 4*

```python
def gradient_descent_runner(points,starting_b,starting_w,learning_rate,num_iteration):
    b=starting_b
    w=starting_w
    cost_history = np.zeros(num_iteration)
    for i in range(num_iteration):
        b,w=step_gradent(b,w,np.array(points),learning_rate)
        print("Iteration Numero :",i)
        print("Fonction du cout = ", compute_error_for_line_given_point(b,w,np.array(points)))
        cost_history[i] = compute_error_for_line_given_point(b,w,np.array(points))
        print("w=", w)
        print("b=", b)
    plt.plot(range(num_iteration), cost_history)
    plt.show()
    return [b,w]
```

## *Question 5*

```python
# Hyperparameter : learning_rate = 0.0001 et nombre_itration = 100 :
def run():

    # Initialize one hundred points
    x=np.linspace(1,10,100)
    y=np.linspace(1,10,100)

    array=[]

    for i in np.arange(100):
        for j in np.arange(100):
            p=[x[i],y[i]]
            array.append(p)


    learning_rate=0.0001
    init_b=0
    init_w=0
    num_iteration=100

    [b,w]=gradient_descent_runner(array,init_b,init_w,learning_rate,num_iteration)

    plt.figure()

    plt.scatter(x, y)
    y2=b+w*x

    plt.scatter(x,y2)
    print("last_b=", b)
    print("last_w=", w)

if __name__=="__main__":
    run()
```

```python
# Importer les librairies : numpy, matplotlib, sklearn.
import numpy as np  # Importation de la bibliothque numpy pour les calculs numriques
import matplotlib.pyplot as plt  # Importation de la bibliothque matplotlib pour la visualisation des donnes
from sklearn import datasets  # Importation de la fonction make_regression pour gnrer des donnes de rgression
from sklearn.linear_model import SGDRegressor  # Importation du modle de rgression linaire avec descente de gradient stochastique (SGD)
from sklearn.model_selection import train_test_split  # Importation de la fonction pour diviser les donnes en ensembles d'entranement et de test
```

## *Question 2*

```python
# Crer votre dataset en important la fonction datasets.make_regression et utilisez la pour gnrer un problme de rgression alatoire de 100 exemples avec une seule variable.
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
# X est la variable explicative, y est la variable cible
# n_samples est le nombre d'exemples  gnrer
# n_features est le nombre de variables explicatives
# noise est le niveau de bruit dans les donnes gnres
# random_state est une graine alatoire pour la reproductibilit des rsultats
```

## *Question 3*

```python
# Visualiser vos donnes en utilisant la fonction plt.scatter(x, y).
plt.scatter(X, y)  # Tracer un nuage de points pour visualiser les donnes
plt.title('Donnes de rgression alatoires')  # Titre du graphe
plt.xlabel('X')  # tiquette de l'axe des abscisses
plt.ylabel('Y')  # tiquette de l'axe des ordonnes
plt.show()  # Afficher le graphe
```

## *Question 4*

```python
# Dfinir votre modle avec SGDRegressor() sur 100 itrations avec un Learning rate de 0.0001.
model = SGDRegressor(max_iter=100, eta0=0.0001, random_state=42)
# Cration d'une instance du modle de rgression linaire avec SGD
# max_iter est le nombre maximal d'itrations
# eta0 est le taux d'apprentissage initial (learning rate)
```

## *Question 5*

```python
# Entraner votre modle en utilisant la mthode fit().
model.fit(X, y)# Entranement du modle sur les donnes d'entranement X avec les valeurs cibles y
```

## *Question 6*

```python
# Calculer la prcision de votre modle en utilisant la fonction score qui calcule le coefficient de dtermination entre le modle et les valeurs y de votre Dataset.
score = model.score(X, y)
# Calcul de la prcision du modle en comparant les prdictions du modle avec les valeurs cibles y
print("Prcision du modle (Coefficient de dtermination) :", score)
```

## *Question 7*

```python
# Faire de nouvelles prdictions avec la fonction predict() et tracer les rsultats avec la fonction plt.plot().
X_test = np.linspace(min(X), max(X), 100).reshape(-1, 1)
# Gnrer de nouvelles donnes pour les prdictions en crant un ensemble de points uniformment rpartis entre les valeurs min et max de X
# reshape(-1, 1) est utilis pour transformer les donnes en une matrice avec une seule colonne
y_pred = model.predict(X_test)
# Faire des prdictions sur les nouvelles donnes gnres
plt.scatter(X, y)  # Tracer les donnes d'entranement
plt.plot(X_test, y_pred, color='red')  # Tracer la ligne de rgression prdictive
plt.title('Rgression linaire avec SGD (100 itrations)')  # Titre du graphe
plt.xlabel('X')  # tiquette de l'axe des abscisses
plt.ylabel('Y')  # tiquette de l'axe des ordonnes
plt.show()  # Afficher le graphe
```

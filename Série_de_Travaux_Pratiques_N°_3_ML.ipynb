## *Question 2*

```python
#
import numpy as np
import matplotlib.pyplot as plt

def compute_error_for_line_given_point(b,w,points):
    totalerror=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        totalerror+=(y-(w*x+b))**2
    return totalerror/(2*float(len(points)))
```

## *Question 3*

```python
#
def step_gradent(b_current,w_current,points,learnrate):
    b_gradient=0
    w_gradient=0
    N=float(len(points))

    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        #
        b_gradient+=(1/N)*((w_current*x+b_current)-y)#
        w_gradient+=(1/N)*x*((w_current*x+b_current)-y)
    #renew
    new_b=b_current-(learnrate*b_gradient)
    new_w=w_current-(learnrate*w_gradient)
    return [new_b,new_w]
```

## *Question 4*

```python
def gradient_descent_runner(points,starting_b,starting_w,learning_rate,num_iteration):
    b=starting_b
    w=starting_w
    cost_history = np.zeros(num_iteration)
    for i in range(num_iteration):
        b,w=step_gradent(b,w,np.array(points),learning_rate)
        print("Iteration Numero :",i)
        print("Fonction du cout = ", compute_error_for_line_given_point(b,w,np.array(points)))
        cost_history[i] = compute_error_for_line_given_point(b,w,np.array(points))
        print("w=", w)
        print("b=", b)
    plt.plot(range(num_iteration), cost_history)
    plt.show()
    return [b,w]
```

## *Question 5*

```python
# Hyperparameter : learning_rate = 0.0001 et nombre_itération = 100 :
def run():

    # Initialize one hundred points
    x=np.linspace(1,10,100)
    y=np.linspace(1,10,100)

    array=[]

    for i in np.arange(100):
        for j in np.arange(100):
            p=[x[i],y[i]]
            array.append(p)


    learning_rate=0.0001
    init_b=0
    init_w=0
    num_iteration=100

    [b,w]=gradient_descent_runner(array,init_b,init_w,learning_rate,num_iteration)

    plt.figure()

    plt.scatter(x, y)
    y2=b+w*x

    plt.scatter(x,y2)
    print("last_b=", b)
    print("last_w=", w)

if __name__=="__main__":
    run()
```

## *Question 1*

```python
# Importer les librairies : numpy, matplotlib, sklearn.
import numpy as np  # Importation de la bibliothèque numpy pour les calculs numériques
import matplotlib.pyplot as plt  # Importation de la bibliothèque matplotlib pour la visualisation des données
from sklearn import datasets  # Importation de la fonction make_regression pour générer des données de régression
from sklearn.linear_model import SGDRegressor  # Importation du modèle de régression linéaire avec descente de gradient stochastique (SGD)
from sklearn.model_selection import train_test_split  # Importation de la fonction pour diviser les données en ensembles d'entraînement et de test
```

## *Question 2*

```python
# Créer votre dataset en important la fonction datasets.make_regression et utilisez la pour générer un problème de régression aléatoire de 100 exemples avec une seule variable.
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
# X est la variable explicative, y est la variable cible
# n_samples est le nombre d'exemples à générer
# n_features est le nombre de variables explicatives
# noise est le niveau de bruit dans les données générées
# random_state est une graine aléatoire pour la reproductibilité des résultats
```

## *Question 3*

```python
# Visualiser vos données en utilisant la fonction plt.scatter(x, y).
plt.scatter(X, y)  # Tracer un nuage de points pour visualiser les données
plt.title('Données de régression aléatoires')  # Titre du graphe
plt.xlabel('X')  # Étiquette de l'axe des abscisses
plt.ylabel('Y')  # Étiquette de l'axe des ordonnées
plt.show()  # Afficher le graphe
```

## *Question 4*

```python
# Définir votre modèle avec SGDRegressor() sur 100 itérations avec un Learning rate de 0.0001.
model = SGDRegressor(max_iter=100, eta0=0.0001, random_state=42)
# Création d'une instance du modèle de régression linéaire avec SGD
# max_iter est le nombre maximal d'itérations
# eta0 est le taux d'apprentissage initial (learning rate)
```

## *Question 5*

```python
# Entraîner votre modèle en utilisant la méthode fit().
model.fit(X, y)# Entraînement du modèle sur les données d'entraînement X avec les valeurs cibles y
```

## *Question 6*

```python
# Calculer la précision de votre modèle en utilisant la fonction score qui calcule le coefficient de détermination entre le modèle et les valeurs y de votre Dataset.
score = model.score(X, y)
# Calcul de la précision du modèle en comparant les prédictions du modèle avec les valeurs cibles y
print("Précision du modèle (Coefficient de détermination) :", score)
```

## *Question 7*

```python
# Faire de nouvelles prédictions avec la fonction predict() et tracer les résultats avec la fonction plt.plot().
X_test = np.linspace(min(X), max(X), 100).reshape(-1, 1)
# Générer de nouvelles données pour les prédictions en créant un ensemble de points uniformément répartis entre les valeurs min et max de X
# reshape(-1, 1) est utilisé pour transformer les données en une matrice avec une seule colonne
y_pred = model.predict(X_test)
# Faire des prédictions sur les nouvelles données générées
plt.scatter(X, y)  # Tracer les données d'entraînement
plt.plot(X_test, y_pred, color='red')  # Tracer la ligne de régression prédictive
plt.title('Régression linéaire avec SGD (100 itérations)')  # Titre du graphe
plt.xlabel('X')  # Étiquette de l'axe des abscisses
plt.ylabel('Y')  # Étiquette de l'axe des ordonnées
plt.show()  # Afficher le graphe
```
